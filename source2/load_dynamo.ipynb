```python
import pandas as pd
```

```python
weather_data = pd.read_csv('..\data_weather\climate_data.csv') # loading data into dataframe
weather_data.head() 
```

```python
from decimal import Decimal
import json
import boto3
```

```python
weather_data = weather_data.rename(columns={'Date': 'date'}) # transforming column names for matching with Dtnadb tamble rowkey

dic_data = weather_data.to_dict('records') # Convert dataframe into a list o json object

ddb_data = json.loads(json.dumps(dic_data), parse_float=Decimal) # transforming float into decimal for using boto3 put_item() 

```

```python
import boto3
s3 = boto3.resource('s3')

dynamodb = boto3.resource('dynamodb')
table = dynamodb.Table('weather_data')
for element in ddb_data:
    #print(element)
    table.put_item(Item=element)
```
